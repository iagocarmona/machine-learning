```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
```

```python
y = x ** 2
```

```python
plt.plot(x,y)
plt.xlabel("$x$")
plt.ylabel("$x^2$")
plt.title("$f(x) = x^2$")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel("$x$")
plt.ylabel("$x^2$")
plt.title("$f(x) = x^2$")

plt.subplot(1,2,2)
plt.plot(x,x**3)
plt.xlabel("$x$")
plt.ylabel("$x^3$")
plt.title("$f(x) = x^3$")

plt.tight_layout()
```

# Modo Orientado a Objetos

```python
fig, ax = plt.subplots(1)
ax.plot(x,y)
ax.set_xlabel("$x$")
ax.set_ylabel("$x^2$")
ax.set_title("$f(x) = x^2$")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].scatter(x,y)
ax[0].set_xlabel("$x$")
ax[0].set_ylabel("$x^2$")
ax[0].set_title("$f(x) = x^2$")

ax[1].plot(x,x**3)
ax[1].set_xlabel("$x$")
ax[1].set_ylabel("$x^3$")
ax[1].set_title("$f(x) = x^3$")

fig.tight_layout()
```

```python
import math

dados = [
    { 'ylabel': "$x^2$", 'title': "$f(x) = x^2$", 'y': x**2},
    { 'ylabel': "$x^3$", 'title': "$f(x) = x^3$", 'y': x**3},
    { 'ylabel': "$x^4$", 'title': "$f(x) = x^4$", 'y': x**4},
    { 'ylabel': "$x^5$", 'title': "$f(x) = x^5$", 'y': x**5},
    { 'ylabel': "$x^6$", 'title': "$f(x) = x^6$", 'y': x**6},
]

total_rows = math.ceil(len(dados)/2)

fig, ax = plt.subplots(nrows=total_rows, ncols=2, figsize=(8,3*total_rows))

for idx, dado in enumerate(dados):

  col = idx % 2
  row = idx // 2

  ax[row][col].plot(x,dado['y'])
  ax[row][col].set_xlabel("$x$")
  ax[row][col].set_ylabel(dado['ylabel'])
  ax[row][col].set_title(dado['title'])

fig.tight_layout()
```

```python
fig, ax = plt.subplots(1)
ax.plot(x, x**2, label="$x^2$", c='red')
ax.plot(x, x**3, label="$x^3$", c='green')
ax.set_xlabel("$x$")
ax.set_ylabel("$f(x)$")
ax.set_title("$x^2$ e $x^3$")
ax.legend(loc="best")

plt.show()
```
